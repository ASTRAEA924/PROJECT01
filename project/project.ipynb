```python
#請使用非colab的IDE來運行(例如:Visual Studio,Anaconda Jupyter Notebook)因為colab沒辦法執行播放音檔等功能(colab是在雲端server上跑程式)
#載入需要使用的套件
import pygame,os  #pygame這個套件需先安裝
from random import randint
#初始化pygame
pygame.init()
#初始化音訊
pygame.mixer.init()

#球反彈功能 
def bbounce():
    global scP1,scP2
    if ball.rect.x >= 790:
        scP1 += 1
        if choose =='sp':    
            ball.velocity = [randint(-14,-8),randint(-8,8)]
        else:
            ball.velocity = [randint(-7,-4),randint(-8,8)]

        ball.rect.x = 395
        ball.rect.y = 195
        s2.play()
    if ball.rect.x <= 0:
        scP2 += 1
        if choose =='sp':    
            ball.velocity = [randint(8,14),randint(-8,8)]
        else:
            ball.velocity = [randint(4,7),randint(-8,8)]
        ball.rect.x = 395
        ball.rect.y = 195
        s2.play()
    if ball.rect.y > 490:
        ball.velocity[1] = -ball.velocity[1]
        s3.play()
    if ball.rect.y < 0:
        ball.velocity[1] = -ball.velocity[1]     
        s3.play()
#選音樂功能
def backmusic():
    x = input('請選擇音樂: \n1.音樂01\n2.音樂02\n3.音樂03\n4.In The End\n5.Faded \n任意鍵.隨機音樂\n')#讓使用者輸入東西來選音樂
    while True:
        if x == '1':
            pygame.mixer.music.load(os.path.join('pong', 'soundtrack01.mp3')) #讀取音樂檔
            return
        elif x == '2':
            pygame.mixer.music.load(os.path.join('pong', 'soundtrack02.mp3')) #讀取音樂檔
            return
        elif x == '3':
            pygame.mixer.music.load(os.path.join('pong', 'soundtrack03.mp3')) #讀取音樂檔
            return
        elif x == '4':
            pygame.mixer.music.load(os.path.join('pong', 'In_The_End.mp3')) #讀取音樂檔
            return
        elif x == '5':
            pygame.mixer.music.load(os.path.join('pong', 'Faded.mp3')) #讀取音樂檔
            return
        else:
            x = str(randint(1,5)) #將x指定為隨機一個歌曲代號

#選擇顏色功能
def color_choose(x):
    #讓使用者透過輸入來選擇顏色
    C = input('請選擇%s的顏色: 1.WHITE 2.PINK 3.VIOLET 4.TIFFANNY 5. CALIFORNIA BLUE 6. LIMEGREEN 7. ORANGE OTHERKEY.隨機顏色:\n'%(x))
    #觸發彩蛋判斷
    if x == '球' and C == '大龜':
        color = 'sp'
    else:    
        if C == '1':
            color = WHITE
        elif C == '2':
            color = PINK
        elif C == '3':
            color = VIOLET
        elif C == '4':
            color = TIFFANY
        elif C == '5':
            color = CALIFORNIA_BLUE
        elif C == '6':
            color = LIMEGREEN
        elif C == '7':
            color = VIVID_ORANGE
        else:        
            color = (randint(0,255),randint(0,255),randint(0,255))#指定隨機的參數給color
    return color    


# 定義常用的顏色(R,G,B)
BLACK = (0,0,0)
WHITE = (255,255,255)
VIOLET = (204,103,255)
PINK = (255,168,238)
TIFFANY = (107,255,230)
CALIFORNIA_BLUE = (108,172,228)
LIMEGREEN = (17,255,0)
VIVID_ORANGE = (255,179,0)
RANDOMC = (randint(0,255),randint(0,255),randint(0,255))
#匯入彩蛋圖片
spimg = pygame.image.load(os.path.join('pong','IMG_2530.jpg'))
#拍子 物件
class Paddle(pygame.sprite.Sprite):
    
    #定義物件外觀
    def __init__(self, color, width, height):
        #初始化
        super().__init__()
        
        #設定拍子圖案的參數
        self.image = pygame.Surface([width, height])
        self.image.fill(BLACK)
        self.image.set_colorkey(BLACK)
 
        # 畫出球拍
        pygame.draw.rect(self.image, color, [0, 0, width, height])
        
        # 回傳物件位置
        self.rect = self.image.get_rect()
    #物件往上移功能    
    def moveUp(self, pixels):
        self.rect.y -= pixels
        #避免跑出視窗
        if self.rect.y < 0:
          self.rect.y = 0
    #物件往下移動功能      
    def moveDown(self, pixels):
        self.rect.y += pixels
        #避免跑出視窗
        if self.rect.y > 425:
          self.rect.y = 425 

#球 物件
class Ball(pygame.sprite.Sprite):
    
    #定義物件外觀
    def __init__(self, color, width, height):
        #初始化
        super().__init__()
        
        
        # 設定物件背景色
        self.image = pygame.Surface([width, height])
        self.image.fill(BLACK)
        self.image.set_colorkey(BLACK)
 
        # 劃出球
        pygame.draw.rect(self.image, color, [0, 0, width, height])
        #球的初速
        self.velocity = [randint(4,8),randint(-8,8)]
        
        # 回傳物件位置
        self.rect = self.image.get_rect()
    #物件位置移動功能    
    def update(self):
        self.rect.x += self.velocity[0]
        self.rect.y += self.velocity[1]
    #物件反彈      
    def bounce(self):
        self.velocity[0] = -(self.velocity[0] + randint(0,2))
        self.velocity[1] = randint(-10,10)
#彩蛋物件
class spBall(pygame.sprite.Sprite):
    
    
    def __init__(self, color, width, height):
        
        super().__init__()
        
        #將球指定為匯入的圖片
        self.image = pygame.transform.scale(spimg,(width,height))
        
        #球的初速
        self.velocity = [randint(6,12),randint(-10,10)]
        
        # 位置
        self.rect = self.image.get_rect()
    #物件位置移動功能    
    def update(self):
        self.rect.x += self.velocity[0]
        self.rect.y += self.velocity[1]
    #物件反彈功能      
    def bounce(self):
        self.velocity[0] = -(self.velocity[0] + randint(0,2))
        self.velocity[1] = randint(-10,10)

#音效匯入
s1 = pygame.mixer.Sound(os.path.join('pong','pong1.mp3'))
s2 = pygame.mixer.Sound(os.path.join('pong','pong2.mp3'))
s3 = pygame.mixer.Sound(os.path.join('pong','pong3.mp3'))

#backmusic()

#pygame.mixer.music.load(os.path.join('pong', 'In_The_End.mp3'))
#pygame.mixer.music.set_volume(0.4)

#參數
PADSP = 8

#背景音樂
#pygame.mixer.music.play(-1)


#P1的拍子 
choose = color_choose('P1的拍子')
P1 = Paddle(choose, 10, 80)
P1.rect.x = 120
P1.rect.y = 200

#P2的拍子
choose = color_choose('P2的拍子')
P2 = Paddle(choose, 10, 80)
P2.rect.x = 670
P2.rect.y = 200

#球
choose = color_choose('球')
if choose == 'sp':  #觸發彩蛋
    ball = spBall(choose,60,60)
    pygame.mixer.music.load(os.path.join('pong', '大龜彩蛋.mp3'))
else:
    ball = Ball(choose,10,10)
    backmusic()
#球的初始位置    
ball.rect.x = 395
ball.rect.y = 195
#播放背景音樂
pygame.mixer.music.set_volume(0.4)
pygame.mixer.music.play(-1)

# 遊戲視窗
SIZE = (800, 500)
screen = pygame.display.set_mode(SIZE)
pygame.display.set_caption("PONG.DEMO") #定義視窗標題

#建立物件LIST
all_sprites_list = pygame.sprite.Group()
 
# 將物件加入物件list中
all_sprites_list.add(P1,P2,ball)

 
# 使主程式持續運行
running = True
 
# 用於控制FPS
clock = pygame.time.Clock()
 
#初始化分數
scP1 = 0
scP2 = 0



# *主程式*
while running:
    # 判定是否關閉遊戲
    for event in pygame.event.get(): #玩家做的指令
        if event.type == pygame.QUIT: # 玩家按下關閉視窗
              running = False # 將running改為False以結束遊戲
        elif event.type==pygame.KEYDOWN:#玩家有沒有按鍵盤
                if event.key==pygame.K_ESCAPE: #按下ESC會結束遊戲
                     running=False
 
    #玩家操控的指令
    keys = pygame.key.get_pressed()
    if keys[pygame.K_w]:
        P1.moveUp(PADSP)
    if keys[pygame.K_s]:
        P1.moveDown(PADSP)
    if keys[pygame.K_UP]:
        P2.moveUp(PADSP)
    if keys[pygame.K_DOWN]:
        P2.moveDown(PADSP)    
 
    #更新物件動態
    all_sprites_list.update()
    
    #檢查球的碰撞
    bbounce()
    #球與拍子碰撞偵測
    if pygame.sprite.collide_rect(ball, P1) or pygame.sprite.collide_rect(ball, P2):
      ball.bounce()
      s1.play()
    
    #畫面更新段落
    #背景
    screen.fill(BLACK)
    #中線
    pygame.draw.line(screen, WHITE, [399, 0], [399, 500], 5)
    
    #畫上物件
    all_sprites_list.draw(screen) 
 
    #顯示分數:
    font = pygame.font.Font(None, 74)
    txt = font.render(str(scP1), 1, WHITE)
    screen.blit(txt, (250,10))
    txt = font.render(str(scP2), 1, WHITE)
    screen.blit(txt, (520,10))
 
    #更新畫面
    pygame.display.flip()
     
    # 設定FPS
    clock.tick(60)
#結束程式
pygame.quit()
```
